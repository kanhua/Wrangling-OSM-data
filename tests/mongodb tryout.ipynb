```python
len(db.london.distinct("address.postcode"))
```

```python
dd=db.london.find({"address.postcode":{"$regex":"^(E|EC|N|NW|SE|SW|W|WC)[0-9][0-9]*\s*[0-9][A-Z][A-Z]"}},{"address.postcode":1})
for ddd in dd:
    print(ddd)
```

```python
db.london.distinct("address.postcode_area")
```

```python
dd=db.london.aggregate([{"$group":{"_id":"$address.postcode_area","count":{"$sum":1}}},
                       {"$sort":{"count":-1}}])
for ddd in dd:
    print(ddd)
```

```python
db.london.distinct("created.time_s.year")
```

```python
dd=db.london.aggregate([{"$group":{"_id":"$created.time_s.year","count":{"$sum":1}}},
                       {"$sort":{"count":-1}}])
for ddd in dd:
    print(ddd)
```
